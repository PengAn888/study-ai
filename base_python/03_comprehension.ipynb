# 列表推导试
*[out_exp_res for out_exp in input_list if conditaion ]*

```python
list=['abc','bcdefg','xyz']

out_exp = [elem for elem in list if len(elem)>3]

print(out_exp)
```

# 字典推导试

*{key_expr:value_expr for value in collection if condition}*


```python
list = {'中国','日本','韩国','美国'}

comprehension = {index:elem for (index,elem) in enumerate(list)}
print(comprehension)
```
