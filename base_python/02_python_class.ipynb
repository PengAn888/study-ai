```python
class MyClass :
    age = 34
    def __init__(self, name):
        self.name = name
    def displayName(self):
        print(self.name)

    
    def displayAge():
        print(MyClass.age)

    # @staticmethod
    def displayAgeStatic():
        print(MyClass.age)
        

an = MyClass('anpeng')

an.displayName()
an.displayAge()
MyClass.displayAgeStatic()

```
