```python
%%writefile ./base_python/anp.py

anp_v = 10
def anp_add(anpList):
    anp_sum = 0
    for i in range(len(anpList)):
        anp_sum += anpList[i]
    return anp_sum


anpList = [1,2,3,4,5]
print(anp_add(anpList))
```
