函数测试

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(0,6,0.1)
y = np.sin(x)
```

```python
print(x)
```

```python
print(y)
```

```python
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('y = sin(x)')
plt.show()
```

```python
# y1 导函数
y1 = np.cos(x)
print(y1)
```

```python
plt.plot(x, y1)
plt.xlabel('x')
plt.ylabel('y')
plt.title('y = cos(x)')
plt.show()
```

```python
# 将函数和导函数的图像放到一起
plt.plot(x, y, label = 'y = sin(x)')
plt.plot(x, y1, label ='y\'=cos(x)', linestyle='--')
plt.xlabel('x')
plt.ylabel('y')
plt.title('y=sin(x)&y\'=cos(x)')
plt.legend()
plt.show()
```
