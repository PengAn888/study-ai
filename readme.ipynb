### 学习路径建议

1. 先通官方入门教程，掌握张量、自动微分、模型训练基础；

2. 用 d2l.ai 或莫烦教程完成 2-3 个小项目（如 MNIST 分类、简单 CNN）；

3. 进阶学习 Effective PyTorch 和官方示例，优化代码性能；

4. 通过 Transformers/PyTorch Geometric 开展专项任务实战；

5. 遇到问题优先查阅 PyTorch Discuss 和 Stack Overflow。
